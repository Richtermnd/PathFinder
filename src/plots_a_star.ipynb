```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('../results/a_star.csv')
df.head()
```

```python
a_star = df[df["name"] == 'a_star']
sides = a_star["side"]
a_star_path = a_star["path"]
a_star_visited = a_star["visited"]
a_star_explored = a_star["explored"]

a_star_boosted = df[df["name"] == 'a_star_boosted']
a_star_boosted_path = a_star_boosted["path"]
a_star_boosted_visited = a_star_boosted["visited"]
a_star_boosted_explored = a_star_boosted["explored"]
```

```python
plt.figure(facecolor="#C0C0C0")
plt.title('Path Length')
plt.plot(sides, a_star_path, label='a_star', linestyle='dashed')
plt.plot(sides, a_star_boosted_path, label='a_star_boosted', linestyle='solid')
plt.legend()
```

```python
plt.figure(facecolor="#C0C0C0")
plt.title('Visited')
plt.xlabel("Сторона местности")
plt.ylabel("Кол-во посещенных узлов")
plt.plot(sides, a_star_visited, label='a_star', linestyle='dashed')
plt.plot(sides, a_star_boosted_visited, label='a_star_boosted', linestyle='solid')
plt.legend()
```

```python
plt.figure(facecolor="#C0C0C0")
plt.title('Explored')
plt.xlabel("Сторона местности")
plt.ylabel("Кол-во изученных узлов")
plt.plot(sides, a_star_explored, label='a_star', linestyle='dashed')
plt.plot(sides, a_star_boosted_explored, label='a_star_boosted', linestyle='solid')
plt.legend()
```
