```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('../surface.csv')
df.head()
```

```python
wave = df[df["name"] == "wave"]
dijkstra = df[df["name"] == "dijkstra"]
a_star = df[df["name"] == "a_star"]
```

```python
sides = wave["side"]
wave_path = wave["path"]
wave_visited = wave["visited"]
wave_explored = wave["explored"]
```

```python
dijkstra_path = dijkstra["path"]
dijkstra_visited = dijkstra["visited"]
dijkstra_explored = dijkstra["explored"]
```

```python
a_start_path = a_star["path"]
a_start_visited = a_star["visited"]
a_start_explored = a_star["explored"]
```

```python
plt.figure(facecolor="#C0C0C0")
plt.title("Path length")
plt.xlabel("Сторона местности")
plt.ylabel("Длина пути")
plt.plot(sides, wave_path, label='wave', linestyle='dotted')
plt.plot(sides, dijkstra_path, label='dijkstra', linestyle='dashed')
plt.plot(sides, a_start_path, label='a_star', linestyle='solid')
plt.legend()
```

```python
plt.figure(facecolor="#C0C0C0")
plt.title("explored")
plt.xlabel("Сторона местности")
plt.ylabel("Кол-во изученных узлов")
plt.plot(sides, wave_explored, label='wave', linestyle='dotted')
plt.plot(sides, dijkstra_explored, label='dijkstra', linestyle='dashed')
plt.plot(sides, a_start_explored, label='a_star', linestyle='solid')
plt.legend()
```

```python
plt.figure(facecolor="#C0C0C0")
plt.title("visited")
plt.xlabel("Сторона местности")
plt.ylabel("Кол-во посещенных узлов")
plt.plot(sides, wave_visited, label='wave', linestyle='dotted')
plt.plot(sides, dijkstra_visited, label='dijkstra', linestyle='dashed')
plt.plot(sides, a_start_visited, label='a_star', linestyle='solid')
plt.legend()
```
